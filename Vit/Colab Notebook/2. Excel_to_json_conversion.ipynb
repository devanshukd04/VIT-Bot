```python
import pandas as pd

data = pd.read_csv("https://docs.google.com/spreadsheets/d/1K4BRxil6ZERLb7QiUmmWBiqU_pD0mp5g1kHxms87Zuc/export?format=csv")
data
```

```python
data = data[['No','Domain','Question',"Answers"]]
```

```python
data = data.dropna()
```

```python
dataset = {"intents":[]}

for index, row in data.iterrows():
  tag = row['Domain'] + row['No']
  question = row['Question'].split(",")
  answer = row['Answers']

  new_dict = {"intent":tag,"text":question,"responses":[answer]}
  dataset["intents"].append(new_dict)

dataset
```

```python
import json

json_object = json.dumps(dataset)
  
# download the data in json format
with open("/content/Json_data.json", "w") as outfile:
    outfile.write(json_object)
```
